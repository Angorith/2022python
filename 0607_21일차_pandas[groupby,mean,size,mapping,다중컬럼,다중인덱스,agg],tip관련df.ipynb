```python
tips = pd.read_csv('https://compmath.korea.ac.kr/appmath/data/tips.csv')
tips
```

```python
tips['tip_pct']= tips['tip']/tips['total_bill']
```

```python
tips.head()
```

map,apply,agg

```python
grouped = tips.groupby(['day','smoker'])
```

```python
grouped['tip_pct'].agg('mean')  #aggregate 여러개의 함수를 동시에 쓰고싶을떄
```

```python
grouped['tip_pct'].agg(['mean','std'])  #std 표준편차 : 평균에서 얼마나 멀어져 있는지(오차) 알아보는 수치
```

```python
grouped['tip_pct'].agg([('평균','mean'),('표준편차','std')])    #count, mean, max, min, std
```

```python
grouped['tip_pct'].agg([('평균','mean'),('표준편차',np.std)])
```

```python
functions = ['count','mean', 'max']
result = grouped['tip_pct','total_bill'].agg(functions)
```

```python
result
```

```python
tips.groupby(['day','smoker'], as_index=False).mean()
```
