```python
pwd = input('비밀번호를 입력하시오')
if len(pwd) < 4:
  print('비밀번호는 4자 이상 입니다.')
else:
  if pwd.isalpha():         #영문입력
    alpha_count=True
    print('영문을 포함해서 입력해주세요.')
  else: 
    if pwd in ['!','@','#','$','%','^','&','*','~']:
      special_count = True
      print('특수문자를 포함해서 입력해주세요')
    else:
     print(f'비밀번호는 {pwd} 입니다')
```

```python
import re

pwd = input('비밀번호:')
numbers = re.sub(r'[^0-9]','',pwd)
strings = re.sub(r'[^a-z]','',pwd)
unique_string = re.sub(r'[^!,@,#,$,%,^,&,*]','',pwd)

if numbers.count('')-1 ==0:
  print('숫자를 포함')
if strings.count('')-1 ==0:
  print('문자를 포함')
if unique_string.count('')-1 ==0:
  print('특문을 포함')
if len(pwd)<4:
  print('4글자 이상')
elif len(pwd)>=4:
  print(f'비밀번호{pwd}')
```

```python
pwd = 'q1w2#'
number_of_character=0
alpha_count = False
numeric_count=False
special_count=False
letters_count = False

if len(pwd)<4:
  print('다시입력해주세요')
for i  in pwd :
  if i.isalpha():
    number_of_character+=1
    alpha_count=True
  if i.isnumeric():
    number_of_character+=1
    numeric_count = True
  if i in ['!','@','#','$','%','^','&','*']:
    number_of_character+=1
    special_count = True

letters_count = (number_of_character ==len(pwd))
codition = alpha_count + numeric_count + special_count + letters_count

print(f'검사한 문자수 : {number_of_character},전체 문자수 : {len(pwd)}')

if codition == 4:
  print(f'비밀번호 {pwd}는 올바른 비밀번호 입니다.')
else:
  print('false')
```
