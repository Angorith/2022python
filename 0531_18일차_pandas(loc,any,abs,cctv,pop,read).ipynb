```python
root_url = 'http://compmath.korea.ac.kr/appmath/data/'
```

```python
names = ['movie_id','title','genres']
```

```python
movies = pd.read_table(root_url + 'movies.dat',sep='::',header = None, names = names, engine='python')
```

sep='::'
각 데이터의 구분이 ::로 구분되어 있다

Toy Story(1995)::Animation|Children's|Comedy

```python
movies[:10]
```

```python
all_genres = []
for x in movies.genres:
    all_genres.extend(x.split('|')) #list로 반환
```

```python
genres = pd.unique(all_genres)  #중복값 제거
```

```python
len(genres)
```

```python
zero_matrix = np.zeros((len(movies), len(genres)))
```

```python
a = np.zeros((len(movies), len(genres)))
```

```python
dummies = pd.DataFrame(zero_matrix, columns=genres)
```

```python
dummies
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key') #left , right에 공통으로 존재하는 것만 뺴서 출력
```

```python
pd.merge(left, right, how='left', on='key') #두개의 기준 left컬럼 기준으로
```

```python
pd.merge(left, right, how='right', on='key')    #없는 값은 NaN으로 표시
```

```python
pd.merge(left, right, how='outer', on='key')    # 합집합
```

```python
pd.merge(left, right, how='inner', on='key')
```
