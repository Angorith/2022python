```python
x= np.arange(1,10,0.1)
loss= lambda x : np.exp(-x)
acc = lambda x : -np.exp(-x)

x1 = np.random.randn(len(x))

fig, loss_ax = plt.subplots(figsize=(8,6))

acc_ax = loss_ax.twinx()    #한 그래프에 축만 다르게 설정

loss_ax.plot(x, loss(x), 'y',label='train loss')    #손실 함수,  train: 훈련
loss_ax.plot(x, loss(x-x1/5),'r', label='val loss') # val : 검증

acc_ax.plot(x, acc(x), 'b',label='train acc')   #훈련
acc_ax.plot(x, acc(x-x1/5), 'g', label= 'val acc')  #검증
for label in acc_ax.get_yticklabels():  #y 축 tick 색깔 지정    #오른쪽 숫자 0.0, 0.1등
    label.set_color('blue')

loss_ax.set_xlabel('epoch',size = 20) #epoch 반복횟수
loss_ax.set_ylabel('loss', size=20) #
acc_ax.set_ylabel('accuracy', color='blue', size=20)    #정확도

loss_ax.legend(loc = 'upper right')
acc_ax.legend(loc='lower right')

plt.show()
```

```python
x= np.arange(0,10,np.pi/100)    #파이값 3.141592 를 100으로 나눈값의 간격으로 x값을 만들겠다.
f = lambda x : np.sin(x)+ x/10  
```

```python
plt.figure()
```

```python
#r = 문자 그대로 보여줘라 , 정규식 표현
#$사이는 수식$
XTN = [r'$0 \pi$',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$',r'$5\pi/2$',r'$3\pi$']
plt.figure(figsize=(8,6))

plt.plot(x, f(x))   #데이터가 들어가는 곳

plt.title('Plot Exercise', position=(0.5,1.04), fontsize=20)    #positon (x값, y값) 기울기순

plt.xlabel(r'$x$', fontsize=20) #수식 x 
plt.ylabel(r'$f (x) = sin(x) + x$', fontsize = 20)  # 라텍스 표현

#tick = 눈금
plt.xticks(np.arange(0.0,10.0,np.pi/2), labels = XTN, fontsize=15)  #np.arange(tick 간격, 이름 xtn)
plt.yticks(np.arange(-1,2.1,np.pi/5),fontsize=15)

plt.text(0.8, -0.3, 'Test Messages', color = 'k',fontsize=18)   #-0.3 = 0 , k = black   #값기준

plt.grid()
plt.tight_layout()  #가로값과 세로값 맞추게 하기
plt.show()
```

```python
event = {'태권도 시작':(8,1),'바이올린 시작':(13,2),  '대학 졸업':(25,-2), '군입대':(26,-3), '현재':(29,0)}
text= []
x=[]
y=[]
for i in event:
    text.append(i)
    x.append(event[i][0])
    y.append(event[i][1])
```

```python
text, x, y
```

```python
plt.figure(figsize=(16,8))
plt.grid()
plt.scatter(x,y)
```

```python
#한글설정
from matplotlib import font_manager
font_path = 'IropkeBatangM.woff'
fontprop = font_manager.FontProperties(fname=font_path, size=15)

plt.figure(figsize=(16,8))
plt.grid()
plt.scatter(x,y)

for idx , age in enumerate(x):
    if y[idx]>0:
        ymin=0
        ymax=y[idx]
        y_position= y[idx]+0.2
    else:
        ymin = y[idx]
        ymax = 0
        y_position = y[idx] - 0.3
    plt.text(age, y_position, text[idx],fontproperties=fontprop, size=15)
    plt.vlines(age, ymin, ymax, color='red',linestyles='--', linewidth=2)
plt.hlines(0,0,60, color='b', linestyle='solid', linewidth=2)
plt.hlines(0,60,100, colors='b', linestyles='--',linewidth=2)
plt.xticks(np.arange(0,100,10))
plt.yticks(np.arange(-4,4,1))
plt.xlim(0,100)
plt.ylim(-3,3)
plt.xlabel('나이', fontproperties=fontprop)
plt.ylabel('행복감',fontproperties=fontprop)
```
